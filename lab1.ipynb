**–ú–µ—Ç–∞ —Ä–æ–±–æ—Ç–∏**:¬†–Ω–∞–±—É—Ç–∏ –≤–º—ñ–Ω–Ω—è —Ç–∞ –¥–æ—Å–≤—ñ–¥ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è –º–µ—Ç–æ–¥—ñ–≤
—Ä–æ–∑–≤‚Äô—è–∑–∞–Ω–Ω—è —á–∞—Å—Ç–∫–æ–≤–æ—ó —Ç–∞ –ø–æ–≤–Ω–æ—ó –ø—Ä–æ–±–ª–µ–º–∏ –≤–ª–∞—Å–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω—å. –ó–∞–∫—Ä—ñ–ø–∏—Ç–∏
—Ç–∞ –ø–æ–≥–ª–∏–±–∏—Ç–∏ —Ç–µ–æ—Ä–µ—Ç–∏—á–Ω—ñ –∑–Ω–∞–Ω–Ω—è –º–µ—Ç–æ–¥—ñ–≤ —Ä–æ–∑–≤‚Äô—è–∑–∞–Ω–Ω—è —á–∞—Å—Ç–∫–æ–≤–æ—ó —Ç–∞
–ø–æ–≤–Ω–æ—ó –ø—Ä–æ–±–ª–µ–º–∏ –≤–ª–∞—Å–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω—å.

````python
import numpy as np #–ø—ñ–¥–∫–ª—é—á–∏–º–æ –±—ñ–±–ª—ñ–æ—Ç–µ–∫—É –º–∞—Ç—Ä–∏—á–Ω–∏—Ö —Ä–æ–∑—Ä–∞—Ö—É–Ω–∫—ñ–≤
np.set_printoptions(precision=6)
````

# –í–∞—Ä—ñ–∞–Ω—Ç 4
### –ú–∞—Ç—Ä–∏—Ü—è –∑ —Ç–∞–±–ª–∏—Ü—ñ 1:

````python
matrix1 = np.array([
    [ 932,  82,  -578, -519,  593],
    [  82,  834, -350, -100,  374],
    [-578, -350,  834,  352, -570],
    [-519, -100,  352,  653, -253],
    [ 593,  374, -570, -253,  898]               
    ])
````

### –ú–∞—Ç—Ä–∏—Ü—è –∑ —Ç–∞–±–ª–∏—Ü—ñ 2:

````python
matrix2 = np.array([
    [ 863, -758, -380, -208, -294],
    [ -17,  756, -345,  553,  771],
    [-856,  129,  872,  661,  654],
    [ 854,  818,  401,  878, -108],
    [-847,   78,  579,  599,  834]            
    ])
````

# –ó–∞–≤–¥–∞–Ω–Ω—è 1 
–ó–Ω–∞–π—Ç–∏ –Ω–∞–π–±—ñ–ª—å—à–µ —Ç–∞ –≤–ª–∞—Å–Ω–µ —á–∏—Å–ª–æ —Ç–∞ –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–∏–π –≤–ª–∞—Å–Ω–∏–π
–≤–µ–∫—Ç–æ—Ä —Å—Ç–µ–ø–µ–Ω–µ–≤–∏–º –º–µ—Ç–æ–¥–æ–º –∞–±–æ –º–µ—Ç–æ–¥–æ–º —Å–∫–∞–ª—è—Ä–Ω–∏—Ö –¥–æ–±—É—Ç–∫—ñ–≤ –∑ —Ç–æ—á–Ω—ñ—Å—Ç—é
$\varepsilon=0,0001$ (–º–∞—Ç—Ä–∏—Ü—è –∑ —Ç–∞–±–ª–∏—Ü—ñ 1).

1. –ó–≤—ñ—Ç –º–∞—î –º—ñ—Å—Ç–∏—Ç–∏ –∫–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–∏.
2. –î–ª—è –º–µ—Ç–æ–¥—É —Å–∫–∞–ª—è—Ä–Ω–∏—Ö –¥–æ–±—É—Ç–∫—ñ–≤ –Ω–∞ –∫–æ–∂–Ω–æ–º—É –∫—Ä–æ—Ü—ñ –≤–∏–≤–æ–¥–∏—Ç–∏ $\lambda^{(k)} = \dfrac{s^{(k)}}{t^{(k)}}$

### –ú–µ—Ç–æ–¥ —Å–∫–∞–ª—è—Ä–Ω–∏—Ö –¥–æ–±—É—Ç–∫—ñ–≤

````python
eps = 0.0001
A = np.copy(matrix1)
y = np.random.rand(5) #–≤ —è–∫–æ—Å—Ç—ñ –ø–µ—Ä—à–æ–≥–æ –Ω–∞–±–ª–∏–∂–µ–Ω–Ω—è –±–µ—Ä–µ–º–æ –≤–∏–ø–∞–¥–∫–æ–≤–∏–π –≤–µ–∫—Ç–æ—Ä
print("First approximation ", y)

iteration = 0 #–Ω–æ–º–µ—Ä –ø–æ—Ç–æ—á–Ω–æ—ó —ñ—Ç–µ—Ä–∞—Ü—ñ—ó
lmbd = ( A.dot(y) @ y ) / (y @ y) #–¥–µ @ - —Å–∫–∞–ª—è—Ä–Ω–∏–π –¥–æ–±—É—Ç–æ–∫, dot - –º–∞—Ç—Ä–∏—á–Ω–µ –º–Ω–æ–∂–µ–Ω–Ω—è
lmbd_prev = np.inf

while (np.abs(lmbd - lmbd_prev) >= eps):
    y = A.dot(y)
    lmbd_prev = lmbd
    lmbd = ( A.dot(y) @ y ) / (y @ y)
    iteration +=1 
    print(" > Iteration %2d, ùúÜ = %f" % (iteration, lmbd))

print("Approximation of major eigenvalue: ", lmbd)
y = A.dot(y)
x = y / np.linalg.norm(y)
print("Approximation of major eigenvector: ", x)

print("\nResults using numpy library:")
value, vect = np.linalg.eigh(A)
print(value[-1],vect[:,-1])
````

–î–ª—è –ø—ñ–¥—Ç–≤–µ—Ä–¥–∂–µ–Ω–Ω—è –∫–æ—Ä–µ–∫—Ç–Ω–æ—Å—Ç—ñ –æ–±—Ä–∞—Ö—É—î–º–æ –≤–µ–∫—Ç–æ—Ä –Ω–µ–≤'—è–∑–∫–∏

````python
r = A.dot(x) - lmbd*x
print("Residual vector:", r)
print("Norm of residual vector:", np.linalg.norm(r))
````

#### **–í–∏—Å–Ω–æ–≤–∫–∏:**
–ü—Ä–∏ –∑–∞–¥–∞–Ω—ñ–π —Ç–æ—á–Ω–æ—Å—Ç—ñ $\varepsilon = 10^{-4}$ –¥–ª—è –¥–∞–Ω–Ω–æ—ó —Å–∏–º–µ—Ç—Ä–∏—á–Ω–æ—ó –º–∞—Ç—Ä–∏—Ü—ñ –º–µ—Ç–æ–¥ —Å–∫–∞–ª—è—Ä–Ω–∏—Ö –¥–æ–±—É—Ç–∫—ñ–≤ –∑–±—ñ–≥–∞—î—Ç—å—Å—è –¥–æ–≤–æ–ª—ñ —à–≤–∏–¥–∫–æ, 8-12 —ñ—Ç–µ—Ä–∞—Ü—ñ–π –≤ –∑–∞–ª–µ–∂–Ω–æ—Å—Ç—ñ –≤—ñ–¥ –ø–µ—Ä—à–æ–≥–æ –Ω–∞–±–ª–∏–∂–µ–Ω–Ω—è. –û—Å–∫—ñ–ª—å–∫–∏ –º–∞—Ç—Ä–∏—Ü—è —Å–∏–º–µ—Ç—Ä–∏—á–Ω–∞, –æ—Ç—Ä–∏–º–∞–Ω—ñ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏ –±—É–ª–∏ –¥–æ–¥–∞—Ç–∫–æ–≤–æ –ø–µ—Ä–µ–≤—ñ—Ä–µ–Ω—ñ –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é —Ñ—É–Ω–∫—Ü—ñ—ó –∑ –±—ñ–±–ª—ñ–æ—Ç–µ–∫–∏ ```numpy``` $-$ —Ä—ñ–∑–Ω–∏—Ü—è –º—ñ–∂ –≤–ª–∞—Å–Ω–∏–º —á–∏—Å–ª–∞–º–∏, –æ—Ç—Ä–∏–º–∞–Ω–∏–º –º–µ—Ç–æ–¥–æ–º —Å–∫–∞–ª—è—Ä–Ω–∏—Ö –¥–æ–±—É—Ç–∫—ñ–≤ —Ç–∞ —Ñ—É–Ω–∫—Ü—ñ—î—é ```np.linalg.eigh``` —Ç–∞–∫–æ—Ö –Ω–µ –ø–µ—Ä–µ–≤–∏—â—É—î $\varepsilon$.

–ü—Ä–∏ —Ç–æ—á–Ω–æ—Å—Ç—ñ $\varepsilon$ –¥–ª—è –≤–ª–∞—Å–Ω–æ–≥–æ —á–∏—Å–ª–∞, –Ω–æ—Ä–º–∞ –Ω–µ–≤'—è–∑–∫–∏ –¥–ª—è –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–æ–≥–æ –≤–ª–∞—Å–Ω–æ–≥–æ –≤–µ–∫—Ç–æ—Ä–∞ –≤–∏–π—à–ª–∞ –∑–Ω–∞—á–Ω–æ –±—ñ–ª—å—à–æ—é $-$ –ø–æ—Ä—è–¥–∫–∞ $5 \cdot 10^{-2}$. –û—Ç–∂–µ, —è–∫—â–æ –ø–µ—Ä—à–æ—á–µ—Ä–≥–æ–≤–æ—é –∑–∞–¥–∞—á–µ—é —î –∑–Ω–∞—Ö–æ–¥–∂–µ–Ω–Ω—è —Å–∞–º–µ –≤–ª–∞—Å–Ω–æ–≥–æ –≤–µ–∫—Ç–æ—Ä–∞, –¥–æ—Ü—ñ–ª—å–Ω–æ –±—Ä–∞—Ç–∏ –∑–∞ —É–º–æ–≤—É –≤–∏—Ö–æ–¥—É –∑ —Ü–∏–∫–ª—É –∑–º–µ–Ω—à–µ–Ω–Ω—è –Ω–æ—Ä–º–∏ –Ω–µ–≤'—è–∑–∫–∏ $\|Ax - \lambda x\| < \varepsilon$.

–†–æ–∑–≤'—è–∑–∞—Ç–∏ –ø–æ–≤–Ω—É –ø—Ä–æ–±–ª–µ–º—É –≤–ª–∞—Å–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω—å:
1. –º–µ—Ç–æ–¥–æ–º –Ø–∫–æ–±—ñ  (–º–∞—Ç—Ä–∏—Ü—è –∑ —Ç–∞–±–ª–∏—Ü—ñ 1);
2. QR –∞–±–æ LR –º–µ—Ç–æ–¥–æ–º (–º–∞—Ç—Ä–∏—Ü—è –∑ —Ç–∞–±–ª–∏—Ü—ñ 2);

#### –í–∏–º–æ–≥–∏ –¥–æ –∑–≤—ñ—Ç—É (–≤ –µ–ª–µ–∫—Ç—Ä–æ–Ω–Ω–æ–º—É –≤–∏–≥–ª—è–¥—ñ) –¥–ª—è –¥—Ä—É–≥–æ–≥–æ –∑–∞–≤–¥–∞–Ω–Ω—è.
1. –ó–≤—ñ—Ç –º–∞—î –º—ñ—Å—Ç–∏—Ç–∏ –∫–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–∏.
2. –î–ª—è –ø—ñ–¥—Ç–≤–µ–¥–∂–µ–Ω–Ω—è –∫–æ—Ä–µ–∫—Ç–Ω–æ—Å—Ç—ñ —Ä–æ–±–æ—Ç–∏ –º–µ—Ç–æ–¥—É –Ø–∫–æ–±—ñ –≤–∏–≤–µ–¥—ñ—Ç—å  –Ω–∞ –∫–æ–∂–Ω—ñ–π —ñ—Ç–µ—Ä–∞—Ü—ñ—ó —Ç–∞–∫—ñ –≤–µ–ª–∏—á–∏–Ω–∏:
    * –º–∞—Ç—Ä–∏—Ü—è, —â–æ –¥—ñ–∞–≥–æ–Ω–∞–ª—ñ–∑—É—î—Ç—å—Å—è (–ê);
    * $i,j$ - —ñ–Ω–¥–µ–∫—Å–∏ –∫–ª—é—á–æ–≤–æ–≥–æ –µ–ª–µ–º–µ–Ω—Ç—É (–º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ –∑–∞ –º–æ–¥—É–ª–µ–º); –∫—É—Ç–æ–≤—ñ –ø–∞—Ä–∞–º–µ—Ç—Ä–∏ $c, s$; –ø—Ä–∏ —Ü—å–æ–º—É –Ω–µ –∑–∞–±—É–¥—å—Ç–µ –ø–µ—Ä–µ–≤—ñ—Ä–∏—Ç–∏, —â–æ $c^2 + s^2 = 1$;
    * —Å—É–º—É –∫–≤–∞–¥—Ä–∞—Ç—ñ–≤ –¥—ñ–∞–≥–æ–Ω–∞–ª—å–Ω–∏—Ö –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ $\delta$, –Ω–µ–¥—ñ–∞–≥–æ–Ω–∞–ª—å–Ω–∏—Ö –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ $2\omega$, –≤—Å—ñ—Ö –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ ( $\delta + 2\omega = \mathrm{const}$!). –¢–æ–±—Ç–æ, –ø–µ—Ä–µ–≤—ñ—Ä–∏—Ç–∏, —â–æ –Ω–æ—Ä–º–∞ –§—Ä–æ–±–µ–Ω—ñ—É—Å–∞ —Ç–æ–±—Ç–æ $\|A\|_F = \sum\limits^n_{i=1} \sum\limits^n_{j=1} a_{ij}^2$ –Ω–µ –∑–º—ñ–Ω—é—î—Ç—å—Å—è, –ø—Ä–∏—á–æ–º—É $\delta = \sum\limits^n_{i=1} a_{ii}^2$ –∑—Ä–æ—Å—Ç–∞—î –Ω–∞ $2a_{ij}^2$, –∞ $2\omega = 2 \sum\limits^{n-1}_{i=1} \sum\limits^n_{j=i+1} a_{ij}^2$ –∑–º–µ–Ω—à—É—î—Ç—å—Å—è –Ω–∞ $2a_{ij}^2$.
    
3. –î–ª—è –º–µ—Ç–æ–¥—ñ–≤ QR —Ç–∞ LR –≤–∏–≤–æ–¥—å—Ç–µ –º–∞—Ç—Ä–∏—Ü—ñ Q, R –∞–±–æ L, R, –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–æ, –Ω–∞ –∫–æ–∂–Ω—ñ–π —ñ—Ç–µ—Ä–∞—Ü—ñ—ó. 

–î–ª—è –∫–æ–∂–Ω–æ—ó –≤–ª–∞—Å–Ω–æ—ó –ø–∞—Ä–∏ $(\lambda, v)$ —è–∫ –ø—ñ–¥—Ç–≤–µ—Ä–¥–∂–µ–Ω–Ω—è –∫–æ—Ä–µ–∫—Ç–Ω–æ—Å—Ç—ñ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—É –≤–∏–≤–µ—Å—Ç–∏ –Ω–∞ –¥—Ä—É–∫ –∑–Ω–∞—á–µ–Ω–Ω—è –≤–µ–∫—Ç–æ—Ä—É –Ω–µ–≤'—è–∑–∫–∏ $x = Av ‚Äì \lambda v$, –¥–µ $v$ ‚Äì –≤–ª–∞—Å–Ω–∏–π –≤–µ–∫—Ç–æ—Ä, $\lambda$ - –≤–ª–∞—Å–Ω–µ —á–∏—Å–ª–æ. –¢–æ—á–Ω—ñ—Å—Ç—å $ \varepsilon= 10^{-4}$.

–£ –≤–∏—Å–Ω–æ–≤–∫–∞—Ö –ø—Ä–æ–∞–Ω–∞–ª—ñ–∑—É–≤–∞—Ç–∏ –æ–¥–µ—Ä–∂–∞–Ω—ñ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏.


## 2.1 –ú–µ—Ç–æ–¥ –Ø–∫–æ–±—ñ

````python
def upper_nondiag(B):
    """
    (–î–æ–ø–æ–º—ñ–∂–Ω–∞ —Ñ—É–Ω–∫—Ü—ñ—è)
    –¶—è —Ñ—É–Ω–∫—Ü—ñ—è –ø–æ–≤–µ—Ä—Ç–∞—î –º–∞—Ç—Ä–∏—Ü—é, –≤ —è–∫—ñ–π –µ–ª–µ–º–µ–Ω—Ç–∏ –Ω–∞–¥ –≥–æ–ª–æ–≤–Ω–æ—é –¥—ñ–∞–≥–æ–Ω–∞–ª–ª—é
    –¥–æ—Ä—ñ–≤–Ω—é—é—Ç—å –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–∏–º –µ–ª–µ–º–µ–Ω—Ç–∞–º –≤–∏—Ö—ñ–¥–Ω–æ—ó –º–∞—Ç—Ä–∏—Ü—ñ, –∞ —É—Å—ñ —ñ–Ω—à—ñ - 0
    """
    i_idx, j_idx = np.mgrid[0:B.shape[0], 0:B.shape[1]]
    M = np.copy(B)
    M[i_idx>=j_idx] = 0 #—Ä–æ–∑–≥–ª—è–¥–∞—î–º–æ –ª–∏—à–µ –µ–ª–µ–º–µ–Ω—Ç–∏ –Ω–∞–¥ –≥–æ–ª–æ–≤–Ω–æ—é –¥—ñ–∞–≥–æ–Ω–∞–ª–ª—é
    return M

def max_nondiag(B):
    """
    –¶—è —Ñ—É–Ω–∫—Ü—ñ—è –ø–æ–≤–µ—Ä—Ç–∞—î —ñ–Ω–¥–µ–∫—Å –Ω–∞–π–±—ñ–ª—å—à–æ–≥–æ –∑–∞ –º–æ–¥—É–ª–µ–º –Ω–µ–¥—ñ–∞–≥–æ–Ω–∞–ª—å–Ω–æ–≥–æ –µ–ª–µ–º–µ–Ω—Ç–∞
    —Å–∏–º–µ—Ç—Ä–∏—á–Ω–æ—ó –º–∞—Ç—Ä–∏—Ü—ñ (B)
    """
    M = np.abs(upper_nondiag(B)) #–º–∞—Ç—Ä–∏—Ü—è –º–æ–¥—É–ª—ñ–≤ –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –≤–µ—Ä—Ö–Ω—å–æ–≥–æ —Ç—Ä–∏–∫—É—Ç–Ω–∏–∫–∞
    return np.unravel_index(np.argmax(M), M.shape)

def sqrsum_nondiag(B):
    """
    –¶—è —Ñ—É–Ω–∫—Ü—ñ—è –ø–æ–≤–µ—Ä—Ç–∞—î —Å—É–º–º—É –∫–≤–∞–¥—Ä–∞—Ç—ñ–≤ –≤–µ—Ä—Ö–Ω—ñ—Ö –Ω–µ–¥—ñ–∞–≥–æ–Ω–∞–ª—å–Ω–∏—Ö –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –º–∞—Ç—Ä–∏—Ü—ñ.
    """
    M = upper_nondiag(B)**2 #–ø—ñ–¥–Ω–µ—Å–µ–Ω–Ω—è –∫–æ–∂–Ω–æ–≥–æ –µ–ª–µ–º–µ–Ω—Ç—É –≤ –∫–≤–∞–¥—Ä–∞—Ç
    return M.sum()

def sign(a):
    """
    –ú–æ–¥–∏—Ñ—ñ–∫–æ–≤–∞–Ω–∞ —Ñ—É–Ω–∫—Ü—ñ—è —Å—ñ–≥–Ω—É–º.
    –í—ñ–¥—Ä—ñ–∑–Ω—è—î—Ç—å—Å—è —Ç–∏–º, —â–æ –ø–æ–≤–µ—Ä—Ç–∞—î 1 –ø—Ä–∏ –∞==0, —Ç–∏–º —Å–∞–º–∏–º –¥–æ–∑–≤–æ–ª—è—é—á–∏ 
    –Ω–µ –≤—Ç—Ä–∞—Ç–∏—Ç–∏ –∑–∞–≥–∞–ª—å–Ω—ñ—Å—Ç—å –≤ –æ—Å–Ω–æ–≤–Ω–æ–º—É —Ü–∏–∫–ª—ñ –ø—Ä–æ–≥—Ä–∞–º–∏.
    """
    if a>=0: return 1
    return -1
````

````python
eps = 0.0001
iteration = 0
B = np.copy(matrix1)
V = np.eye(B.shape[0]) #–º–∞—Ç—Ä–∏—Ü—è, —â–æ –≤ –∫—ñ–Ω—Ü—ñ –±—É–¥–µ –º—ñ—Å—Ç–∏—Ç–∏ –≤–ª–∞—Å–Ω—ñ –≤–µ–∫—Ç–æ—Ä–∏

while 2*sqrsum_nondiag(B) >= eps:
    iteration += 1
    print("\n>>> Iteration %d"%iteration)
    print("",B)
                       
    i, j = max_nondiag(B)
    print("Argmax (%d,%d)"%(i,j))

    tau = (B[i,i] - B[j,j]) / (2 * B[i,j])
    t = -tau + sign(tau)*np.sqrt(tau**2 + 1) 
    
    c = 1 / np.sqrt(1 + t**2)
    s = c*t
    print("c =% .10f, s =% .10f, c¬≤+s¬≤ = %f" % (c,s,c**2 + s**2))
    
    delta = (np.diag(B)**2).sum()
    omega = 2*sqrsum_nondiag(B)
    print("ùõø = %f, 2ùúî = %f, sum = %f" % (delta,omega,delta+omega))
    
    T = np.eye(B.shape[0]) #–±—É–¥—É—î–º–æ –º–∞—Ç—Ä–∏—Ü—é –ø–ª–æ—Å–∫–æ–≥–æ –æ–±–µ—Ä—Ç–∞–Ω–Ω—è
    T[i,i] = T[j,j] = c
    T[i,j] = -s
    T[j,i] = s
    
    V = V @ T        # @ –≤ –¥–∞–Ω–Ω–æ–º—É –≤–∏–ø–∞–¥–∫—É —ñ–Ω—Ç–µ—Ä–ø—Ä–µ—Ç—É—î—Ç—å—Å—è —è–∫ –º–∞—Ç—Ä–∏—á–Ω–µ –º–Ω–æ–∂–µ–Ω–Ω—è
    B = T.T @ B @ T  # .T - –æ–ø–µ—Ä–∞—Ü—ñ—è —Ç—Ä–∞–Ω—Å–ø–æ–Ω—É–≤–∞–Ω–Ω—è

print("\n>>> RESULTS")
print("Matrix B =")
print(B)
print("Eigenvalues:")
print(np.diag(B))
print("Eigenvectors:")
print(V)

print("\nResults using numpy library:")
value, vect = np.linalg.eigh(matrix1)
print(value[::-1])
print(vect[:,::-1])
````

–î–ª—è –ø—ñ–¥—Ç–≤–µ—Ä–¥–∂–µ–Ω–Ω—è –∫–æ—Ä–µ–∫—Ç–Ω–æ—Å—Ç—ñ –æ–±—Ä–∞—Ö—É—î–º–æ –≤–µ–∫—Ç–æ—Ä–∏ –Ω–µ–≤'—è–∑–∫–∏

````python
A = np.copy(matrix1)
R = A@V - np.diag(B)*V #–º–∞—Ç—Ä–∏—Ü—è –≤–µ–∫—Ç–æ—Ä—ñ–≤ –Ω–µ–≤'—è–∑–∫–∏

for index, lmbd in enumerate(np.diag(B)):
    print(">>> %d EIGENVALUE" % (index+1))
    print("(",lmbd,",",V[:,index],")")
    print("Residual:", R[:,index])
    print("Norm of residual vector:", np.linalg.norm(R[:,index]))
    print()

print("Norm of residual MATRIX:", np.linalg.norm(R))
````

#### **–í–∏—Å–Ω–æ–≤–∫–∏**:
–ü—Ä–∏ –¥–∞–Ω–∏—Ö —É–º–æ–≤–∞—Ö –º–µ—Ç–æ–¥ –Ø–∫–æ–±—ñ –∑–±—ñ–≥–∞—î—Ç—å—Å—è –∑–∞ 24 —ñ—Ç–µ—Ä–∞—Ü—ñ—ó, —â–æ —î –¥–æ–≤–æ–ª—ñ —Ö–æ—Ä–æ—à–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–º, –≤—Ä–∞—Ö–æ–≤—É—é—á–∏ –≤—ñ–Ω –¥–∞—î —É—Å—ñ –≤–ª–∞—Å–Ω—ñ –∑–Ω–∞—á–µ–Ω–Ω—è —Ç–∞ –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω—ñ –≤–µ–∫—Ç–æ—Ä–∏, –ø—Ä–∏ —Ü—å–æ–º—É –≤ –≤–∏–ø–∞–¥–∫—É –º–∞—Ç—Ä–∏—Ü—ñ $5\times5$ –Ω–µ–¥—ñ–∞–≥–æ–Ω–∞–ª—å–Ω–∏—Ö –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ 10 –ø–∞—Ä, —Ç–æ–±—Ç–æ –≤ –ø—Ä–æ—Ü–µ—Å—Å—ñ –¥–æ—Å—è–∂–Ω–æ—Å—Ç—ñ —Ä—ñ–≤–Ω—è —Ç–æ—á–Ω–æ—Å—Ç—ñ $\varepsilon = 10^{-4}$ –∫–æ–∂–Ω–∞ –ø–∞—Ä–∞ –º—ñ–Ω—ñ–º—ñ–∑—É—î—Ç—å—Å—è –≤ —Å–µ—Ä–µ–¥–Ω—å–æ–º—É 2-3 —Ä–∞–∑–∏. –ü–µ—Ä–µ–≤—ñ—Ä–∫–∏ –ø–æ–∫–∞–∑–∞–ª–∏, —â–æ –Ω–∞ –∫–æ–∂–Ω—ñ–π —ñ—Ç–µ—Ä–∞—Ü—ñ—ó —Å—É–º–∞ –∫–≤–∞–¥—Ä–∞—Ç—ñ–≤ –∫—É—Ç–æ–≤–∏—Ö –∫–æ–µ—Ñ—ñ—Ü—ñ—î–Ω—Ç—ñ–≤ –¥–æ—Ä—ñ–≤–Ω—é—î –æ–¥–∏–Ω–∏—Ü—ñ, –∞ –Ω–æ—Ä–º–∞ –§—Ä–æ–±–µ–Ω—ñ—É—Å–∞ –º–∞—Ç—Ä–∏—Ü—ñ –Ω–µ –∑–º—ñ–Ω—é—î—Ç—å—Å—è (–¥–æ—Ä—ñ–≤–Ω—é—î 6986563) –∑ –≤–µ–ª–∏–∫–∏–º —Ä—ñ–≤–Ω–µ–º —Ç–æ—á–Ω–æ—Å—Ç—ñ. –î–æ–¥–∞—Ç–∫–æ–≤–∞ –ø–µ—Ä–µ–≤—ñ—Ä–∫–∞ –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é –±—ñ–±–ª—ñ–æ—Ç–µ–∫–∏ ```numpy```, —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏ —è–∫–æ—ó, –≤—Ä–∞—Ö–æ–≤—É—é—á–∏ –Ω–µ–≤–µ–ª–∏–∫–∏–π –∑–∞–¥–∞–Ω–∏–π —Ä—ñ–≤–µ–Ω—å —Ç–æ—á–Ω–æ—Å—Ç—ñ $\varepsilon = 10^{-4}$, –º–æ–∂–µ–º–æ –ø—Ä–∏–π–Ω—è—Ç–∏ –∑–∞ –¥—ñ–π—Å–Ω—ñ –∑–Ω–∞—á–µ–Ω–Ω—è, –ø–æ–∫–∞–∑–∞–ª–∞ —â–æ –æ—Ç—Ä–∏–º–∞–Ω—ñ –º–µ—Ç–æ–¥–æ–º –Ø–∫–æ–±—ñ –≤–ª–∞—Å–Ω—ñ —á–∏—Å–ª–∞ —Ç–∞ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∏ –º–∞—Ç—Ä–∏—Ü—ñ –≤–ª–∞—Å–Ω–∏—Ö –≤–µ–∫—Ç–æ—Ä—ñ–≤ –≤—ñ–¥—Ä—ñ–∑–Ω—è—é—Ç—å—Å—è –≤—ñ–¥ –Ω–∏—Ö –Ω–µ –±—ñ–ª—å—à –Ω—ñ–∂ –Ω–∞ $\varepsilon$.

–î–ª—è –ø–µ—Ä–µ–≤—ñ—Ä–∫–∏ –æ—Ç—Ä–∏–º–∞–Ω–∏—Ö –≤–ª–∞—Å–Ω–∏—Ö —á–∏—Å–µ–ª —Ç–∞ –≤–µ–∫—Ç–æ—Ä—ñ–≤, –±—É–ª–∏ –æ–±—Ä–∞—Ö–æ–≤–∞–Ω—ñ –≤–µ–∫—Ç–æ—Ä–∏ –Ω–µ–≤'—è–∑–∫–∏ –¥–ª—è –∫–æ–∂–Ω–æ—ó –ø–∞—Ä–∏. –ù–æ—Ä–º–∞ –≤–∏—è–≤–∏–ª–∞—Å—å –ø–æ—Ä—è–¥–∫—É $5 \times 10^{-3}$ –¥–ª—è –Ω–∞–π–±—ñ–ª—å—à–æ–≥–æ –≤–µ–∫—Ç–æ—Ä–∞ –Ω–µ–≤'—è–∑–∫–∏, —â–æ –Ω–µ —î –¥—É–∂–µ –≥–∞—Ä–Ω–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–º, –∞–ª–µ, –ø—Ä–∏–Ω–∞–π–º–Ω—ñ, –∫—Ä–∞—â–∏–º –∞–Ω—ñ–∂ –≤ –º–µ—Ç–æ–¥—ñ —Å–∫–∞–ª—è—Ä–Ω–∏—Ö –¥–æ–±—É—Ç–∫—ñ–≤.

## 2.2 QR –º–µ—Ç–æ–¥

````python
A = np.copy(matrix2)

from scipy.linalg import qr

iteration = 0
while sqrsum_nondiag(A.T) >= eps: #–Ω–∏–∂–Ω—ñ–π —Ç—Ä–∏–∫—É—Ç–Ω–∏–∫ = –≤–µ—Ä—Ö–Ω—ñ–π —Ç—Ä–∞–Ω—Å–ø–æ–Ω–æ–≤–∞–Ω–æ—ó
    iteration += 1
    print("\n>>> Iteration %d"%iteration)
    print("A=")
    print(A)
    
    Q, R = qr(A)
    print("Q=")
    print(Q)
    print("R=")
    print(R)
    A = R @ Q
    
print("\n>>> RESULTS")
print("Matrix A = ")
print(A)
print("Eigenvalues:")
print(np.diag(A))

print("\nResults using numpy library:")
value, vect = np.linalg.eig(matrix2)
print(value)
````

#### **–í–∏—Å–Ω–æ–≤–∫–∏**:
–ù–∞ –∑–∞–¥–∞–Ω—ñ–π –º–∞—Ç—Ä–∏—Ü—ñ —Ä—ñ–≤–µ–Ω—å —Ç–æ—á–Ω–æ—Å—Ç—ñ $\varepsilon = 10^{-4}$, –ø—Ä–∏ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—ñ QR-–º–µ—Ç–æ–¥—É –∑ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è–º QR —Ä–æ–∑–∫–ª–∞–¥—É –∑ –±—ñ–±–ª—ñ–æ—Ç–µ–∫–∏ ```scipy```, –¥–æ—Å—è–≥–∞—î—Ç—å—Å—è –∑–∞ 28 —ñ—Ç–µ—Ä–∞—Ü—ñ–π. –î–æ–¥–∞—Ç–∫–æ–≤–æ –ø–µ—Ä–µ–≤—ñ—Ä–∏–≤—à–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏ –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é ```numpy``` (—Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏ —è–∫–æ—ó –≤–≤–∞–∂–∏—Ç–∏–º–µ–º–æ –∑–∞ –¥—ñ–π—Å–Ω—ñ) –º–æ–∂–µ–º–æ –ø–æ–º—ñ—Ç–∏—Ç–∏, —â–æ –ø–æ—Ö–∏–±–∫–∞ –∑–æ—Å–µ—Ä–µ–¥–∂–µ–Ω–∞ –≤ –Ω–∞–π–±—ñ–ª—å—à–∏—Ö –¥–≤–æ—Ö –∑–∞ –º–æ–¥—É–ª–µ–º –≤–ª–∞—Å–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–Ω—è—Ö, —Ä—ñ–∑–Ω–∏—Ü—è —è–∫–∏—Ö –∑ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º–∏ ```numpy``` –º–∞—î –ø–æ—Ä—è–¥–æ–∫ $10^{-2}$, —Ç–æ–¥—ñ —è–∫ —É —ñ–Ω—à–∏—Ö –Ω–µ –ø–µ—Ä–µ–≤–∏—â—É—î $\varepsilon$. –¢–∞–∫–∏–º —á–∏–Ω–æ–º —Ü–µ–π –º–µ—Ç–æ–¥ —î –Ω–∞–π–ø–æ–≤—ñ–ª—å–Ω–∏—à–∏–º –∑ —Ä–æ–∑–≥–ª—è–Ω—É—Ç–∏—Ö, –ø—Ä–æ—Ç–µ —î–¥–∏–Ω–∏–º –∑ –Ω–∏—Ö, —â–æ –º–æ–∂–µ –ø—Ä–∞—Ü—é–≤–∞—Ç–∏ –∑ –Ω–µ—Å–∏–º–µ—Ç—Ä–∏—á–Ω–∏–º–∏ –º–∞—Ç—Ä–∏—Ü—è–º–∏.

# –í–∏—Å–Ω–æ–≤–∫–∏
–ü—ñ–¥ —á–∞—Å –≤–∏–∫–æ–Ω–∞–Ω–Ω—è –ª–∞–±–æ—Ä–∞—Ç–æ—Ä–Ω–æ—ó —Ä–æ–±–æ—Ç–∏ –±—É–ª–æ –Ω–∞ –º–æ–≤—ñ –ø—Ä–æ–≥—Ä–∞–º—É–≤–∞–Ω–Ω—è ```Python``` –∑ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è–º –±—ñ–±–ª—ñ–æ—Ç–µ–∫–∏ –º–∞—Ç—Ä–∏—á–Ω–∏—Ö —Ä–æ–∑—Ä–∞—Ö—É–Ω–∫—ñ–≤ ```numpy``` —Ä–µ–∞–ª—ñ–∑–æ–≤–∞–Ω–æ —Ç—Ä–∏ –º–µ—Ç–æ–¥–∏: –º–µ—Ç–æ–¥ —Å–∫–∞–ª—è—Ä–Ω–∏—Ö –¥–æ–±—É—Ç–∫—ñ–≤ –¥–ª—è –∑–Ω–∞—Ö–æ–¥–∂–µ–Ω–Ω—è —Å—Ç–∞—Ä—à–æ–≥–æ –≤–ª–∞—Å–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–Ω—è —Ç–∞ –≤–µ–∫—Ç–æ—Ä–∞, —Ç–∞ –¥–≤–∞ –º–µ—Ç–æ–¥–∏ —Ä–æ–∑–≤'—è–∑–∞–Ω–Ω—è –ø—Ä–æ–±–ª–µ–º–∏ –≤–ª–∞—Å–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω—å - –º–µ—Ç–æ–¥ –Ø–∫–æ–±—ñ —Ç–∞ QR-–º–µ—Ç–æ–¥. –ù–∞ –º–æ—é –¥—É–º–∫—É, —Å–µ—Ä–µ–¥ —Ä–æ–∑–≥–ª—è–Ω—É—Ç–∏—Ö, –º–µ—Ç–æ–¥ –Ø–∫–æ–±—ñ –≤–∏—è–≤–∏–≤—Å—è –Ω–∞–π–±—ñ–ª—å—à —à–≤–∏–¥–∫–∏–º —Ç–∞ —Ç–æ—á–Ω–∏–º –ø—Ä–∏ –∑–Ω–∞—Ö–æ–¥–∂–µ–Ω–Ω—ñ –≤–ª–∞—Å–Ω–∏—Ö –≤–µ–∫—Ç–æ—Ä—ñ–≤ (–ø—Ä–∏ –æ–¥–Ω–æ–∫–æ–≤—ñ–π —Ç–æ—á–Ω–æ—Å—Ç—ñ –¥–ª—è –≤–ª–∞—Å–Ω–∏—Ö —á–∏—Å–µ–ª), –ø—Ä–æ—Ç–µ –≤—ñ–Ω –ø—Ä–∞—Ü—é—î –ª–∏—à–µ –¥–ª—è —Å–∏–º–µ—Ç—Ä–∏—á–Ω–∏—Ö –º–∞—Ç—Ä–∏—Ü—å, —Ç–∞ –ø–æ—Ç—Ä–µ–±—É—î –±–∞–≥–∞—Ç–æ —Ä–æ–∑—Ä–∞—Ö—É–Ω–∫—ñ–≤, –æ—Ç–∂–µ –π–æ–≥–æ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è –Ω–µ –∑–∞–≤–∂–¥–∏ —î –¥–æ—Ü—ñ–ª—å–Ω–∏–º. –¢–∞–∫–æ–∂, —Å–ª—ñ–¥ –∑–∞–∑–Ω–∞—á–∏—Ç–∏, —â–æ –æ–∫—Ä—ñ–º —Ç–æ—á–Ω–æ—Å—Ç—ñ —Å–∞–º–∏—Ö –∞–ª–≥–æ—Ä–∏—Ç–º—ñ–≤ –Ω–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏ –≤–ø–ª–∏–Ω—É–ª–∏ —Ç–æ—á–Ω—ñ—Å—Ç—å –∫–æ–º–ø—å—é—Ç–µ—Ä–Ω–æ–≥–æ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–Ω—è —á–∏—Å–µ–ª —Ç–∞ –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–æ –∫–æ–º–ø—å—é—Ç–µ—Ä–Ω–∏—Ö —Ä–æ–∑—Ä–∞—Ö—É–Ω–∫—ñ–≤.
